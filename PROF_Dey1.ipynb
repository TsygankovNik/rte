```python
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import drive
import os
import sys
directory_path = "/content/drive/MyDrive/practicePicList"
if os.path.exists(directory_path):
    print(" .")
else:
    print("  .  .")
jpg_images = [image for image in os.listdir(path="//content/drive/MyDrive/practicePicList") if image.endswith('.jpg')]
for jpg_image in jpg_images:
    try:
        img = Image.open(os.path.join(directory_path, jpg_image))  #  
        plt.imshow(img)  #  
        plt.show()  #  
        new_name = jpg_image.split('.')[0] + '.png'
        img.save(new_name)  #     
        print(' ')
    except IOError as error:
        print('Couldn\'t read {} '.format(jpg_image))
```
